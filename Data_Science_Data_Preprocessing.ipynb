1. DATA CLEANING

1.1 Missing Values Handling

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,5))
plt.subplot(141)
sns.histplot(df.Age, kde=True)

plt.subplot(142)
sns.boxplot(df.Age)
plt.show()
```

Nilai ini tidak ada outlernya

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(strategy='median')
df.Age = imp.fit_transform(np.array(df.Age).reshape(-1,1))[:,0]
df.Age
```

```python
imp = SimpleImputer(strategy='mean')
df.Salary = imp.fit_transform(np.array(df.Salary).reshape(-1,1))[:,0]
df.Salary
```

2.2 Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cat = ['Country', 'Purchased']

for i in cat:
  le = LabelEncoder()
  df[i]=le.fit_transform(df[i])

df
```

2.3 Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

ndf = scaler.fit_transform(df)
ndf = pd.DataFrame(ndf, columns=df.columns)
ndf.head()
```
