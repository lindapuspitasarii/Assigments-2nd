## 1. Data Cleaning

## 1.1 Missing Values Handling

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(16,5))
plt.subplot(141)
sns.histplot(df.Pregnancies	, kde=True)

plt.subplot(142)
sns.boxplot(df.Pregnancies	)

plt.subplot(143)
sns.histplot(df.Glucose, kde=True)

plt.subplot(144)
sns.boxplot(df.Glucose)

plt.subplot(146)
sns.histplot(df.BloodPressure, kde=True)

plt.subplot(147)
sns.boxplot(df.BloodPressure)

plt.subplot(148)
sns.histplot(df.Age, kde=True)

plt.subplot(149)
sns.boxplot(df.Age)
plt.show()
```

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(strategy='median')
df.Age = imp.fit_transform(np.array(df.Age).reshape(-1,1))[:,0]
df.Age
```

```python
imp = SimpleImputer(strategy='mean')
df.Salary = imp.fit_transform(np.array(df.Salary).reshape(-1,1))[:,0]
```

## 2.2 Encoding

```python
# before encoding
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

cat = ['Country', 'Purchased']

for i in cat:
  le = LabelEncoder()
  df[i] = le.fit_transform(df[i])

df
```

## 4. Resampling

```python
ndf.Purchased.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=21)
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_res, y_res = sm.fit_resample(X,y)
```

```python
df_res = X_res.copy()
df_res['Purchased'] = y_res

df_res.Purchased.value_counts()
```
